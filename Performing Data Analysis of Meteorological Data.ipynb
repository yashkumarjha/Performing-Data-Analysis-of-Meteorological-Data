```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('F:/Data set Python/Performing data analysis using Meterological data/weatherHistory.csv')
```

```python
weather.head()
```

```python
titles_req = ['Formatted Date','Apparent Temperature (C)','Humidity']
df = weather[titles_req]
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df_1 = df.set_index('Formatted Date')
df_1 = df_1.resample('MS').mean()
```

```python
df_1.head()
```

```python
plt.figure(figsize = (14,6))
plt.title("Variation in Apparent temperature and humidity with time.")
plt.plot(df_1)
```

```python
df_april=df_1[df_1.index.month == 4]
```

```python
plt.figure(figsize=(14,6))
plt.title("Plot of average temperature and humidity of the month of April over 10 years.")
plt.plot(df_april)
```
